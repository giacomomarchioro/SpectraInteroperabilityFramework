```python
import json
import numpy
import matplotlib.pyplot as plt
```

```python
with open('json_data.json') as f:
    data = json.load(f)
```

```python
data['proccesedValue']['data'][0]
```

```python
fig, ax = plt.subplots()
ax.plot(data['variablesLabels']['data'][0],
        data['proccesedValue']['data'][0]
        )
ax.set_ylim(data['proccesedValue']['expectedMinValue'],data['proccesedValue']['expectedMaxValue']*1.1)
ax.set_xlabel(f"{data['variablesLabels']['name']} ({data['variablesLabels']['unit']})")
ax.set_ylabel(f"{data['proccesedValue']['type']} ({data['proccesedValue']['unit']})")
```
